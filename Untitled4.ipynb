As we see, only the release year is in integer format. Can we describe it? Also, from the previous analysis, we can see that there are many null or missing values in the data, so we need to treat these missing values before we proceed.

```python
data.describe()
```

First, we change the 'date_added' column. As we see, it is a datetime object, and for easier manipulation, we need to convert it to datetime format. This makes it much easier to handle for further analysis.

```python
data['date_added'] = pd.to_datetime(data['date_added'].astype('string'),format='mixed')
```

```python
data.info()
```

Now, we also need to handle the missing values. First, we identify which columns have missing values and count them. Based on this, we can decide on the best method to impute the null values, whether to treat them or drop them.

```python
data.isnull().sum()
```

As we can see, the duration, rating, and date_added columns have the least number of null values. We can check these and drop the null values. In this case, if we check the rows where duration is null, we can see that the value for duration has been mistakenly added to the rating column. Since the number of affected rows is small and we can easily spot the mistake, we can change the values from the rating column to the duration column for the null values. For the date_added column, we will simply drop the rows with null values.

```python
data[data['duration'].isnull()]
```

```python
d_values = data[data['duration'].isnull()]['rating']
```

```python
data.loc[data[data['duration'].isnull()].index,'rating']= np.nan
```

```python
data['duration']=data['duration'].fillna(d_values)
```

For the categorical data value columns, we can either treat them with the mode, median, or mean, as dropping them would result in a significant loss of data. In this case, we will fill the null values with the 'Unknown' value.
We can also treat the 'director' column by taking the mode with respect to the country and cast. However, if we check the data, there are still some rows where all three columns are missing, so rather than trying to impute them, we will simply fill the missing values directly with 'Unknown'.

```python
data[['cast','country','director']].isnull().value_counts().reset_index()
```

```python
data['director'] = data['director'].fillna('Unknown')
data['cast'] = data['cast'].fillna('Unknown')
data['country'] = data['country'].fillna('Unknown')
data['rating'] = data['rating'].fillna('Unknown')
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

For the duration column, we can also change it to integer type, as we only need to extract the numerical value from the column. In case we need to analyze it, we can use the 'type' column to check which value it corresponds to.

```python
data['duration'] = data['duration'].str.split(' ').str[0].astype(int)
```

Now that we have treated all the missing values and changed the variable types, we need to unnest the different columns that contain a list of data/names. For example, the 'Country' column contains all the countries where the movie/TV show is produced.


---

First, we create a deepcopy of the data, as we need the original unnested data for analysis as well, and unnest the data into a new dataframe.

After this, we need to perform two main steps:

1.   Split the data values in the columns by ','.
2.   Then, we need to explode/expand those values across the entire dataset.

To improve filtering, it's best to first split by ', ' (with a space after the comma) and then explode/expand the data. Afterward, split again by ',' and expand the data once more.

Once we're done with unnesting, we need to drop the extra index column to clean the data.


```python
un_nested_data = data.copy()
```

```python
un_nested_data['cast'] = un_nested_data['cast'].str.split(',')
un_nested_data['director'] = un_nested_data['director'].str.split(',')
un_nested_data['country'] = un_nested_data['country'].str.split(',')
un_nested_data['listed_in'] = un_nested_data['listed_in'].str.split(',')
```

```python
un_nested_data = un_nested_data.explode('cast')
un_nested_data = un_nested_data.explode('director')
un_nested_data = un_nested_data.explode('country')
un_nested_data = un_nested_data.explode('listed_in')
```

```python
un_nested_data.reset_index(drop=True, inplace=True)
```
